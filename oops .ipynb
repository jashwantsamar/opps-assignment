```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
In object-oriented programming (OOP), a class is a blueprint or a template for creating objects. It defines the attributes (data) and behaviors (methods) that objects of that class can have. An object, on the other hand, is an instance of a class. It is a concrete representation of a particular entity that has its own state and behavior.

For example, let's consider a class called Person that describes a person with a name and age. Here's an example of how to define the class in Python:

python
Copy code
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def say_hello(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
In this class, we have two attributes: name and age, which are defined in the constructor method __init__(). We also have a method called say_hello() that prints out a greeting using the person's name and age.

To create an object of this class, we can simply call the constructor and pass in the necessary parameters:

python
Copy code
person1 = Person("Alice", 25)
person2 = Person("Bob", 30)
Here, person1 and person2 are two separate instances of the Person class, each with their own name and age attributes. We can call the say_hello() method on each object to get a personalized greeting:

python
Copy code
person1.say_hello()  # Output: "Hello, my name is Alice and I am 25 years old."
person2.say_hello()  # Output: "Hello, my name is Bob and I am 30 years old."
In this example, the Person class serves as a template for creating multiple objects, each with its own unique set of attributes and behaviors.




JASHWANT SAMAR
Name the four pillars of OOPs.
The four pillars of Object-Oriented Programming (OOP) are:

Encapsulation: Encapsulation refers to the bundling of data and methods that act on that data within a single unit, i.e., a class. This concept provides data hiding and prevents outside code from accessing and modifying an object's internal state directly.

Abstraction: Abstraction involves the separation of essential features of an object from its implementation details, enabling the creation of simplified, higher-level models that can be used to solve complex problems.

Inheritance: Inheritance allows a class to inherit properties and methods from a parent class, enabling code reuse and promoting code organization.

Polymorphism: Polymorphism refers to the ability of objects of different classes to be used interchangeably, allowing for more flexibility and extensibility in object-oriented design. This can be achieved through method overriding or method overloading.




JASHWANT SAMAR
Q3. Explain why the __init__() function is used. Give a suitable example.
In object-oriented programming (OOP), the __init__() function is a special method that is called when an object is created. It is used to initialize the object's attributes and define its initial state. The __init__() method is called a constructor method because it constructs or creates the object.

Here's an example that demonstrates the use of __init__():

python
Copy code
class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.odometer_reading = 0
    
    def get_description(self):
        desc = f"{self.year} {self.make} {self.model}"
        return desc.title()
    
    def read_odometer(self):
        print(f"This car has {self.odometer_reading} miles on it.")
    
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back the odometer!")
    
    def increment_odometer(self, miles):
        self.odometer_reading += miles
In this example, we have defined a Car class with four attributes: make, model, year, and odometer_reading. The __init__() method initializes the make, model, and year attributes using the values passed in as arguments. It also sets the odometer_reading attribute to 0 by default.

We also have several other methods that allow us to interact with a Car object, such as get_description(), which returns a formatted string representing the car's make, model, and year, and update_odometer() and increment_odometer(), which allow us to update the car's odometer reading.

To create an instance of a Car object, we can call the class constructor and pass in the required arguments:

python
Copy code
my_car = Car("toyota", "camry", 2022)
In this example, we have created a Car object called my_car with the make "Toyota", model "Camry", and year 2022. The __init__() method is called automatically when we create the object, initializing the object's attributes to their default values. We can then use the object's other methods to interact with it and perform various operations.




JASHWANT SAMAR
Q4. Why self is used in OOPs?
In object-oriented programming (OOP), self is a reference to the current object that is being manipulated or operated on. It is a special variable that is passed as the first argument to class methods and is used to access an object's attributes and methods within that method.

In OOP, we define classes to create objects that have their own properties and behaviors. When we create an instance of a class, we can access and modify its attributes using dot notation. For example, if we have a Person class with a name attribute, we can create an instance of that class and set its name like this:

python
Copy code
class Person:
    def __init__(self, name):
        self.name = name

person1 = Person("Alice")
person1.name = "Bob"  # change the person1's name to Bob
In this example, we are accessing the name attribute of the person1 object using dot notation. However, when we define a class method, we need to use self to refer to the object being operated on. For example, if we wanted to define a say_hello() method for the Person class, we could do it like this:

python
Copy code
class Person:
    def __init__(self, name):
        self.name = name
    
    def say_hello(self):
        print(f"Hello, my name is {self.name}.")

person1 = Person("Alice")
person1.say_hello()  # Output: "Hello, my name is Alice."
In this example, the say_hello() method takes a single argument, self, which is a reference to the person1 object. Inside the method, we use self.name to access the object's name attribute and print a greeting using that name.

In summary, self is used in OOP to refer to the current object being operated on inside class methods. It allows us to access an object's attributes and methods and manipulate its state within the context of a method.




JASHWANT SAMAR
What is inheritance? Give an example for each type of inheritance.
Inheritance is a fundamental concept in object-oriented programming that allows us to create new classes based on existing ones. It is a way to reuse code and promote code organization by creating hierarchies of related classes.

Inheritance is achieved by defining a new class that inherits the attributes and methods of an existing class, which is called the parent or superclass. The new class is called the child or subclass. The child class can add its own attributes and methods or override the ones inherited from the parent class.

There are four types of inheritance:

Single inheritance: In single inheritance, a subclass inherits from a single superclass. It is the simplest form of inheritance.
Example:

python
Copy code
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Dog(Animal):
    def bark(self):
        print("Woof!")

my_dog = Dog("Max", 3)
print(my_dog.name)  # Output: "Max"
In this example, we have a parent class called Animal that has two attributes: name and age. We have also defined a child class called Dog that inherits from the Animal class using the syntax class Dog(Animal):. The Dog class has an additional method called bark(). We can create instances of the Dog class that have access to the name and age attributes of the parent Animal class.

Multiple inheritance: In multiple inheritance, a subclass inherits from multiple superclasses. It allows the child class to combine the attributes and methods of multiple parent classes.
Example:

python
Copy code
class Car:
    def start_engine(self):
        print("Engine started.")

class Radio:
    def play_music(self):
        print("Music playing.")

class SmartCar(Car, Radio):
    pass

my_car = SmartCar()
my_car.start_engine()  # Output: "Engine started."
my_car.play_music()  # Output: "Music playing."
In this example, we have defined two parent classes: Car and Radio. The Car class has a method called start_engine(), and the Radio class has a method called play_music(). We have also defined a child class called SmartCar that inherits from both Car and Radio using the syntax class SmartCar(Car, Radio):. The SmartCar class does not have any additional attributes or methods of its own. We can create instances of the SmartCar class that have access to the methods of both parent classes.

Hierarchical inheritance: In hierarchical inheritance, multiple child classes inherit from a single parent class. It is a way to create a hierarchy of related classes.
Example:

python
Copy code
class Animal:
    def breathe(self):
        print("Breathing...")

class Mammal(Animal):
    def give_birth(self):
        print("Giving birth...")

class Reptile(Animal):
    def lay_eggs(self):
        print("Laying eggs...")

class Cat(Mammal):
    def meow(self):
        print("Meowing...")

class Snake(Reptile):
    def hiss(self):
        print("Hissing...")

my_cat = Cat()
my_cat.breathe()  # Output: "Breathing..."
my_cat.give_birth()  # Output: "Giving birth..."
my_cat.meow()  # Output: "Meowing..."
In this example, we have defined a parent class called Animal that has a method called breathe(). We have also defined two child classes that inherit from Animal: Mammal and 
```
